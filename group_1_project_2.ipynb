```python
import random
# Name: Olaoluwa Adedamola


#Create a dictionary name as Survey
survey = {}

# SINCE ITS A SAMPLE SURVEY, WE CHOSE TO GENERATE RANDOM NAMES AND HOBBIES

# CREATING A POOL FOR THE FIRST AND LAST NAMES
first_names = ["Emma", "Liam", "Olivia", "Noah", "Ava", "James", "Sophia", "Lucas",
                "Mia", "Ethan","Kweku", "Payton", "Isabella", "Elijah", "Ola",
                 "Ase", "Malinka", "Harold", "Logan", "Charlotte", "Benjamin", "Aria", "Daniel"
                ]

last_names = ["Smith", "Brown", "Johnson", "Taylor", "Anderson", "Davis",
               "Moore", "Wilson", "Thomas", "Clark", "Damola", "Lopez", "Hassan",
              "Lewis", "Robinson", "Walker", "Hall", "Allen", "Young", "King", "Wright", "Scott", "Green"
            ]

# CREATING A POOL OF DIFFERENT HOBBIES
hobby_pool = [
        "Reading", "Gaming", "Cooking", "Swimming", "Hiking",
        "Soccer", "Drawing", "Basketball", "Talking", "Singing",
        "Dancing", "Watching Movies", "Painting", "Knitting", "Editing",
        "Content Creating", "Coding", "Skating", "Cycling", "Swimming",
        "Working-Out", "Traveling", "Baking", "Photography", "Gardening"
            ]

# LOOP TO GEERATE NAMES AND ASSIGN RANDOMLY, HOBBIES TO EACH OF THEM
while len(survey) < 50:
    first = random.choice(first_names)
    last = random.choice(last_names)
    full_name = f"{first} {last}"

    if full_name in survey:       # Avoid duplicates
        continue

    # ASSIGN RANDOM HOBBIES TO EACH NAME
    num_hobbies = random.randint(2,4)
    selected_hobbies = set(random.sample(hobby_pool, num_hobbies))

    survey[full_name] = selected_hobbies

# PRINT THE FIRST 10 SURVEYS
print("\n---First 10 Survey Entries ---")
count = 0

for name, hobbies in survey.items():
    hobby_str = ", ".join(sorted(hobbies))
    print(f"{name}: {hobby_str}")

    count += 1
    if count == 10:
        break
```

```python
# NAME: OLAOLUWA ADEDAMOLA
# TASK 2 â€“ Analyze the Survey Data

# 1. Find the total number of unique hobbies in the survey
all_hobbies = []

for hobbies in survey.values():
    all_hobbies.extend(hobbies)

unique_hobbies = set(all_hobbies)
print(f"\nTotal Unique Hobbies: {len(unique_hobbies)}")


# 2. Find the most common hobby & how many participants like it
hobby_count = {}

for hobby in all_hobbies:
    hobby_count[hobby] = hobby_count.get(hobby, 0) + 1

most_common = max(hobby_count, key=hobby_count.get)
print(f"Most Common Hobby: {most_common} (liked by {hobby_count[most_common]} participants)")
```

```python
# Name:Kwaku Ameyaw
# Task 3: People with same hobbies
def group_people_by_hobby(survey):
    hobby_groups = {}
    # your code here
    for person, hobbies in survey.items():
      for hobby in hobbies:
        if hobby not in hobby_groups:
          hobby_groups[hobby] = set()
        hobby_groups[hobby].add(person)
    return hobby_groups

# Print each hobby and the list of names of participants who share it.
hobby_groups = group_people_by_hobby(survey)
for hobby, participants in hobby_groups.items():
    print(f"{hobby}: {', '.join(participants)}")
```

```python
# Name: Kwaku Ameyaw
# Task 4: Add entry and update file
name = input("\nEnter new participant's full name: ")
hobbies_input = input("Enter 2 to 4 hobbies, separated by commas: ")
new_hobbies = {h.strip() for h in hobbies_input.split(",")}

if len(new_hobbies) < 2 or len(new_hobbies) > 4:
  print("Invalid entry. Enter bewtween 2 and 4 hobbies.")
else:
  survey[name] = new_hobbies
  print("\nEntry successfully added")

print(f"\nUpdated survey size: {len(survey)} participants")

print(f"{name}: {new_hobbies}")


```
